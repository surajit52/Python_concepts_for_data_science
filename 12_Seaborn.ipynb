# Correlation Based on Heatmap

A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
#distplot is used to univariate analysis
#joinplot is used to bivariate analysis(relationship between 2 variables)
#pairplot is used to do multivariate analysis(ralationship between multiple variables)
```

```python
#tip is dependet feature
#sex,smoker,day etc are independent
```

```python
#heatmap is used to provide correlation(inter relation) between each and every feature
#correlation can only be found for integer or float type
#whenver we apply correlation,value rangea from -1 to +1 called pearson correlation
```

```python
df.dtypes
```

```python
#to find correlation

df.corr()

#	total_bill,tip and size are float dataypes hence correlation can be found for only these three features
```

```python
#heretotal_bill and tip is correlated by a postive factor of 0.67
#which means if total_bill increases, bill also increases
```

```python
sns.heatmap(df.corr())
```

Joinplot actually helps to do univariate/bivariate analysis

```python
sns.jointplot(x='tip',y='total_bill', data= df, kind='hex')

#kind='hex' means the representation will be in form of hexagonal structures
```

```python
sns.jointplot(x='tip',y='total_bill', data= df, kind='reg')
```

# Pair plot

A pair plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

making pairplot with repsect to particular feature using hue='feature'

```python
sns.pairplot(data=df,hue='sex')
```

# Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python

sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about
- boxplot
- violinplot 
- countplot
- bar plot


```python
##Countplot
#here we have give either x-axis or y-axis information, other information is count
sns.countplot('sex', data= df)
```

```python
sns.countplot('day', data= df)
```

```python
sns.countplot(y= 'sex', data= df)
```

```python
#barplot
#both x-axis and y-axis are to beprovided


sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

# Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot('sex','total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill', data=df,palette = 'rainbow')
```

```python
sns.boxplot(data=df,orient='v')

#it is considering all the features 
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

# Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
